```python
import requests
import pandas as pd

BROKER = "http://orionld_orga:1026/ngsi-ld/v1"
HEADERS = {"Accept": "application/ld+json"}

r = requests.get(f"{BROKER}/entities?type=Observation", headers=HEADERS)
data = r.json()
rows = []
for e in data:
    val = e["value"]["value"]
    prop = e["observedProperty"]["value"]
    time_ = e["resultTime"]["value"]
    rows.append({"property": prop, "value": val, "time": time_})

df = pd.DataFrame(rows)
df
```
